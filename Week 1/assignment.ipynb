```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]

df = pd.read_csv(url, names=columns)

print("Dataset imported successfully.")
df.head(5)
```

```python
df_cleaned = df.copy()
# Numerical columns
numeric_columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

# Fill missing numerical values with median
for column in numeric_columns:
    df_cleaned[column] = df_cleaned[column].fillna(df_cleaned[column].median())

# Fill missing categorical values with mode
df_cleaned["species"] = df_cleaned["species"].fillna(df_cleaned["species"].mode()[0])

# Check missing values after imputation
print("\nMissing values after imputation:")
print(df_cleaned.isnull().sum())
```

```python
# Normalize numerical columns to a scale between 0 and 1

numeric_columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

df_normalized = df_cleaned.copy()

df_normalized[numeric_columns] = (
    df_normalized[numeric_columns] - df_normalized[numeric_columns].min()
) / (
    df_normalized[numeric_columns].max() - df_normalized[numeric_columns].min()
)

df_normalized.head(5)
```

```python
# Encoding: Convert categorical species labels into numbers
df_transformed = df_cleaned.copy()
df_transformed["species_code"] = pd.factorize(df_transformed["species"])[0]
df_transformed[["species", "species_code"]].head(5)
```

```python
# OLAP Operation 1: Slicing
from IPython.display import display

slicing_result = df_transformed[df_transformed["species"] == "Iris-setosa"]

print("Slicing Result: Iris-setosa flowers only")
display(slicing_result.head(5))
```

```python
# OLAP Operation 2: Dicing
from IPython.display import display
dicing_result = df_transformed[
    (df_transformed["species"] == "Iris-setosa") &
    (df_transformed["sepal_length"] > 5.0)
]
print("Dicing Result: Iris-setosa flowers with sepal_length greater than 5.0")
display(dicing_result.head(5))
```

```python
# OLAP Operation 3: Roll-up
from IPython.display import display
numeric_columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
rollup_result = df_transformed.groupby("species")[numeric_columns].mean()

print("Roll-up Result: Average measurements grouped by species")
display(rollup_result)
```

```python
# OLAP Operation 4: Drill-down
from IPython.display import display
df_transformed["petal_size_category"] = pd.cut(
    df_transformed["petal_length"],
    bins=[0, 2, 5, 7],
    labels=["Small", "Medium", "Large"]
)
drilldown_result = df_transformed.groupby(
    ["species", "petal_size_category"],
    observed=True
)[numeric_columns].mean()
print("Drill-down Result: Average measurements grouped by species and petal size category")
display(drilldown_result)
```

```python
# Step 1: Data Integration using concat
from IPython.display import display
df_part1 = df.iloc[:75]
df_part2 = df.iloc[75:]

df_integrated_concat = pd.concat([df_part1, df_part2], ignore_index=True)
print("Data Integration using concat completed.")
print("Original dataset shape:", df.shape)
print("Integrated dataset shape:", df_integrated_concat.shape)

display(df_integrated_concat.head(5))
```

```python
# Step 2: Data Integration using merge
species_info = pd.DataFrame({
    "species": ["Iris-setosa", "Iris-versicolor", "Iris-virginica"],
    "flower_group": ["Group A", "Group B", "Group C"]
})
df_integrated_merge = pd.merge(df_integrated_concat, species_info, on="species", how="left")
print("Data Integration using merge completed.")

display(df_integrated_merge.head(5))
```

```python
# Step 3: Feature Selection
selected_features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

X = df_cleaned[selected_features]
y = df_cleaned["species"]

print("Selected features for analysis:")
print(selected_features)

display(X.head(5))
```

```python
# Step 5: Visualization
for species_name in df_cleaned["species"].unique():
    species_data = df_cleaned[df_cleaned["species"] == species_name]
    plt.scatter(
        species_data["petal_length"],
        species_data["petal_width"],
        label=species_name
    )
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width by Iris Species")
plt.legend()
plt.show()
```

```python
# Data Mining: Applying K-Means Clustering Algorithm

from sklearn.cluster import KMeans
from IPython.display import display

# Select numerical columns
numeric_columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

# Make a copy of cleaned data
df_clustered = df_cleaned.copy()

# Make sure numerical columns are numeric
for column in numeric_columns:
    df_clustered[column] = pd.to_numeric(df_clustered[column], errors="coerce")

# Fill any missing numerical values with median
for column in numeric_columns:
    df_clustered[column] = df_clustered[column].fillna(df_clustered[column].median())

# Select features for clustering
X = df_clustered[numeric_columns]

# Create K-Means model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

# Add cluster result to dataset
df_clustered["cluster"] = kmeans.fit_predict(X)

print("K-Means clustering completed successfully.")
display(df_clustered.head(5))
```

```python
# Evaluation / Interpretation

from IPython.display import display

# Show average values for each cluster
cluster_summary = df_clustered.groupby("cluster")[numeric_columns].mean()

print("Cluster Summary:")
display(cluster_summary)

print("Interpretation:")
print("The K-Means algorithm grouped the iris flowers into three clusters.")
print("The cluster summary shows the average measurements for each cluster.")
print("Petal length and petal width are useful features because they help separate the flower groups.")
print("Flowers with smaller petal measurements are separated from flowers with larger petal measurements.")
print("This shows that the Iris dataset has clear patterns based on flower measurements.")
```
