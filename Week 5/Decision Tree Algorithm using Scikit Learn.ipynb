# Decision Tree Algorithm using Scikit Learn

### What is Decision Tree Algorithm?

 Decision tree algorithm is a supervised learning algorithm that can be used for both classification and regression tasks. It creates a tree-like model of decisions and their possible consequences, with each internal node representing a "test" on an attribute, each branch representing the outcome of the test, and each leaf node representing a class label or a value. The algorithm starts with a single node (the root) and recursively splits the data into subsets based on the values of the input features, until it reaches a point where the subsets are "pure" (i.e., contain only samples from a single class or a single value). The result is a tree that can be used to make predictions on new data by traversing the tree from the root to a leaf node.

### Decision Tree Algorithm working process:

The decision tree algorithm has a flowchart-like structure, where each internal node represents a feature(or attribute), each branch represents a decision rule, and each leaf node represents the outcome. The process of building a decision tree can be broken down into the following steps:

- **Select the best feature**: The algorithm starts by selecting the feature that best separates the data based on the target variable. This is typically done using a metric such as information gain or Gini index.

- **Create a decision rule**: A decision rule is created based on the selected feature's value. For example, if the feature is "age" and the decision rule is "age <= 30", then all the samples that have "age <= 30" will be directed to the left branch, while all the others will be directed to the right branch.

- **Create sub-nodes**: The samples are split into subsets based on the decision rule, and new sub-nodes are created for each subset.

- **Repeat**: The process is repeated for each sub-node by selecting the best feature and creating decision rules for the subsets until a stopping criterion is met.

- **Assign classes**: Once the tree is built, it can be used to make predictions by traversing the tree from the root to a leaf node and returning the class or value associated with that leaf node.

It's worth noting that the decision tree algorithm is a **greedy algorithm**, meaning it makes the locally optimal choice at each step. This can sometimes lead to overfitting, where the tree is too complex and performs poorly on new, unseen data. To overcome this, techniques such as pruning or ensemble methods like Random Forest can be used.

### Types of Decision Tree Algorithm:
 There are several types of decision tree algorithms, each with their own specific characteristics and use cases. Some of the most common types include:

- **ID3 (Iterative Dichotomiser 3):** This is one of the earliest decision tree algorithms, developed by Ross Quinlan in 1986. It uses the concept of entropy and information gain to construct the tree.

- **C4.5:** An extension of ID3, C4.5 can handle both continuous and categorical features and can also prune the tree to avoid overfitting.

- **C5.0:** An extension of C4.5, C5.0 can handle missing values, and also uses boosting to improve the accuracy of the tree.

- **CART (Classification and Regression Trees):** This is a decision tree algorithm that can be used for both classification and regression tasks. It uses Gini impurity as a measure of the quality of a split.

- **CHAID (Chi-squared Automatic Interaction Detection):** This decision tree algorithm is used for categorical dependent variables and is particularly useful for market research and customer segmentation.

- **MARS (Multivariate Adaptive Regression Splines):** This algorithm is a non-parametric method for fitting a decision tree, which is particularly useful for high-dimensional data.

- **Random Forest:** It is an ensemble method that creates multiple decision trees and combines their predictions to improve the overall accuracy.

All these decision tree algorithms are different in terms of the way they split the data and the way they overcome overfitting. The choice of algorithm depends on the specific use case, the type of data, and the desired level of interpretability.

### Use a decision tree algorithm when:

- **Clear and interpretable decision-making process:** Decision trees are relatively simple to understand, interpret, and visualize, making them a good choice when the decision-making process needs to be clearly understood.

- **Categorical or numerical variables:** Decision trees can handle both categorical and numerical variables, making them a versatile tool for a wide range of datasets.

- **Identify important features:** Decision trees can help identify the most important features in a dataset, which can be useful for feature selection and feature engineering.

- **Handle missing values:** Decision trees can handle missing values by simply creating a branch for missing values or by imputing values using the most common value or mean.

- **Handle non-linearly separable data:** Decision trees can handle non-linearly separable data by creating a decision boundary using multiple splits, making them a robust algorithm for complex datasets.

- **Reduce overfitting:** Decision trees can be regularized using techniques such as pruning or ensemble methods like Random Forest, which can help prevent overfitting.

It's worth noting that these are general guidelines, and the choice of algorithm depends on the specific use case, the type of data, and the desired level of interpretability. Sometimes an algorithm that is not ideal in theory may work better in practice and vice versa.

### Don't use a decision tree algorithm when:

- **Small amount of data:** Decision trees can easily overfit with small datasets, and in this case, simpler models such as logistic regression or linear regression would be more appropriate.

- **Large number of features:** Decision trees can become very complex when dealing with a large number of features, making them difficult to interpret and prone to overfitting.

- **Need for high precision:** Decision trees are not known for providing very high precision, and in cases where high precision is required, other algorithms such as Random Forest or Boosting methods may be more suitable.

- **Need for smooth decision boundaries:** Decision trees can create jagged and complex decision boundaries, which may not be suitable in cases where smooth decision boundaries are needed, such as in image recognition tasks.

- **Need for online/real-time predictions:** Decision trees can be relatively slow to make predictions, and in cases where real-time predictions are needed, other algorithms such as Random Forest or Boosting methods may be more suitable.

- **Need for low memory usage:** Some decision tree algorithms such as ID3 and C4.5 require to store the tree in memory, and in case you have a limited memory storage, other algorithms such as Random Forest or Boosting methods may be more suitable.

It's worth noting that these are general guidelines, and the choice of algorithm depends on the specific use case, the type of data, and the desired level of interpretability. Sometimes an algorithm that is not ideal in theory may work better in practice and vice versa.

```python
# Importing Important Libraries
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Load the Iris dataset
iris = datasets.load_iris() 
```

When working with Scikit-Learn datasets like iris, the data is stored in a special object called a Bunch (which acts like a Python dictionary). You cannot pass the whole object directly into pd.DataFrame(). 

Before using any data, we must check if the keys we need (data, target, feature_names) actually exist inside the object.

```python
# Check what is inside the object
keys_list = list(iris.keys())
values_list = list(iris.values())

print("keys_list", keys_list)
```

```python
# print(iris.data)
print(len(iris.data))
print(len(iris.target))
print(len(iris.feature_names))
```

To build a successful pandas DataFrame, your matrix of numbers must perfectly align with your list of column names. Let's check their shapes using .shape and len().

```python

# 1. Check rows and columns of the matrix
rows, columns = iris.data.shape
print(f"The matrix has {rows} rows and {columns} columns.")

# 2. Check the number of feature names
num_features = len(iris.feature_names)
print(f"We have {num_features} column names.")

if columns == num_features:
    print("\n SUCCESS: The number of data columns matches the feature names. We CAN create a DataFrame!")
else:
    print("\n ERROR: Dimension mismatch. We cannot create the DataFrame.")
```

```python
df = pd.DataFrame(data=iris.data, columns= iris.feature_names)
df
```

```python
#  Add the target array as a new column named 'target'
df['target'] = iris.target
df.head(100)
```

```python
print(df['target'].unique())
print(df['sepal length (cm)'].unique())
print(df['petal width (cm)'].unique())

print(len(df['sepal length (cm)'].unique()))
print(len(df['petal length (cm)'].unique()))
print(len(df['sepal width (cm)'].unique()))
print(len(df['petal width (cm)'].unique()))
```

```python
# Spliting the data set into X and y
X = iris.data
y = iris.target
```

```python
print(X.shape, X.size)
print(y.shape, y.size)

print(X[:5,])
```

```python
print(y[:500])
```

### Train/Test Split

We **never** train and test on the same data — that's like grading a student on the homework they already saw. We split the dataset:
- **70%** for training (the model learns from this)
- **30%** for testing (we hide this from the model, then check accuracy)

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,      # 30% held out for testing
    random_state=42,     # reproducible split
    stratify=y           # preserve class proportions
)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set:     {X_test.shape[0]} samples")
```

```python
# Create an instance of the decision tree classifier
clf = DecisionTreeClassifier(random_state=42, criterion='entropy')
```

```python
# Fit the classifier to the training data
clf.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = clf.predict(X_test)
```

```python
# Print the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
# Visualize the decision tree
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True);
plt.show()
```

```python
from sklearn.metrics import (
    precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay,  classification_report
)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("--- Standard Metrics ---")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-score:  {f1:.4f}\n")

```

```python
# Comprehensive text report
print("--- Classification Report ---")
print(classification_report(y_test, y_pred))
```

```python
# Generate and plot Confusion Matrix
print("--- Confusion Matrix ---")
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#  Plot the matrix visually
disp = ConfusionMatrixDisplay(confusion_matrix=cm) # Pass display_labels=nb.classes_ if available
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

- **Handling missing values:** Decision trees can handle missing values by simply creating a branch for missing values or by imputing values using the most common value or mean.

- **Handling non-linearly separable data:** Decision trees can handle non-linearly separable data by creating a decision boundary using multiple splits, making them a robust algorithm for complex datasets.

-  Can be used for both classification and regression problems

- Can be used to estimate missing values and maintain accuracy of the model.

- Provide a white-box type of model, in other words, if a given situation is observable in a model, the explanation for the condition is easily explained by Boolean logic.

- Decision trees can handle high dimensional data with good accuracy.

### Disadvantage of Decision Tree Algorithm: 

Decision tree algorithms also have several disadvantages, including:

- **Tendency to overfit:** Decision trees can easily overfit with small datasets or with a large number of features, which can lead to poor generalization performance on unseen data.

- **Not suitable for high precision:** Decision trees are not known for providing very high precision, and in cases where high precision is required, other algorithms such as Random Forest or Boosting methods may be more suitable.

- **Complex decision boundaries:** Decision trees can create jagged and complex decision boundaries, which may not be suitable in cases where smooth decision boundaries are needed, such as in image recognition tasks.

- **Can be sensitive to small changes in the data:** Decision tree algorithms can be sensitive to small changes in the data, and this can make them less robust to noise in the data.

- **Not suitable for online/real-time predictions:** Decision trees can be relatively slow to make predictions, and in cases where real-time predictions are needed, other algorithms such as Random Forest or Boosting methods may be more suitable.

- **Not suitable for low memory usage:** Some decision tree algorithms such as ID3 and C4.5 require to store the tree in memory, and in case you have a limited memory storage, other algorithms such as Random Forest or Boosting methods may be more suitable.

- Can be biased towards features with many levels, this is because the algorithm may select the feature with the most levels as the root node, which can lead to overfitting.

- Decision trees are not efficient when you have a lot of features, it's hard to find the best split among many features.

- Decision trees can be unstable because small variations in the data can result in a completely different tree being generated.

- Decision tree learners create biased trees if some classes dominate. It is therefore recommended to balance the dataset prior to fitting with the decision tree.
