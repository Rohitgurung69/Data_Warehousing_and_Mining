```python
import numpy as np
import pandas as pd

np.random.seed(42)

# keeping a clean backup before adding noise
df_clean_original = df.copy()

# 1. Missing values
random_idx = np.random.choice(df.index, size=10, replace=False)
df.loc[random_idx, "Hours_Studied"] = np.nan

print("Missing values Added to Hours Studied")
display(df.loc[random_idx, ["Hours_Studied"]])
```

Noise 1: Missing Values

I added missing values to the Hours_Studied column. This column was chosen because study hours are important for student performance analysis. This simulates a real-world situation where students may skip a question on a form, or the school system may fail to record their study hours.

```python
# 2. Impossible / out-of-range values - simulate entry errors

df.loc[5, "Attendance"] = -10          # impossible attendance percentage
df.loc[12, "Exam_Score"] = 150         # unrealistic exam score
df.loc[20, "Sleep_Hours"] = 30         # impossible sleep hours in one day

print("----- Out-of-Range Values Added -----")
display(df.loc[[5, 12, 20], ["Attendance", "Exam_Score", "Sleep_Hours"]])
```

```python
# 3. Unrecognized categories - simulate inconsistent data entry

df.loc[3, "Internet_Access"] = "Maybe"       
df.loc[25, "Gender"] = "M/F"                 
df.loc[30, "School_Type"] = "Unknown Type"  

print("----- Unrecognized Category Values Added -----")
display(df.loc[[3, 18, 25], ["Internet_Access", "Gender", "School_Type"]])
```

```python
# 4. Duplicate rows - simulate repeated imports or form resubmissions

df = pd.concat([df, df.iloc[0:5]], ignore_index=True)

print("----- Duplicate Rows Added -----")
print("New dataset shape after adding duplicates:")
print(df.shape)

display(df.tail(10))
```

```python
# Detect missing values
print("----- Missing Values Detection -----")
print(df.isnull().sum())
print("----- Rows with Missing Hours_Studied -----")
display(df[df["Hours_Studied"].isnull()])
```

```python
# Detect impossible / out-of-range values

print("----- Out-of-Range Values Detection -----")
out_of_range = df[
    (df["Attendance"] < 0) | (df["Attendance"] > 100) |
    (df["Exam_Score"] < 0) | (df["Exam_Score"] > 100) |
    (df["Sleep_Hours"] < 0) | (df["Sleep_Hours"] > 24)
]
display(out_of_range[["Attendance", "Exam_Score", "Sleep_Hours"]])
```

```python
# Detect unrecognized categories

print("----- Unrecognized Categories Detection -----")

valid_internet_access = ["Yes", "No"]
valid_gender = ["Male", "Female"]
valid_school_type = ["Public", "Private"]

invalid_categories = df[
    (~df["Internet_Access"].isin(valid_internet_access)) |
    (~df["Gender"].isin(valid_gender)) |
    (~df["School_Type"].isin(valid_school_type))
]

display(invalid_categories[["Internet_Access", "Gender", "School_Type"]])
```

```python
# Detect duplicate rows

print("----- Duplicate Rows Detection -----")

print("Number of duplicate rows:")
print(df.duplicated().sum())

print("----- Duplicate Rows -----")
display(df[df.duplicated()])
```

```python
# Step 5 - Clean the Data
# 1. Fix missing values in Hours_Studied

df["Hours_Studied"] = df["Hours_Studied"].fillna(df["Hours_Studied"].median())

print("----- Missing Values Fixed -----")
print("Missing values in Hours_Studied:", df["Hours_Studied"].isnull().sum())

display(df[["Hours_Studied"]].head(10))
```

```python
# 2. Fix invalid / out-of-range values

df.loc[(df["Attendance"] < 0) | (df["Attendance"] > 100), "Attendance"] = np.nan
df.loc[(df["Exam_Score"] < 0) | (df["Exam_Score"] > 100), "Exam_Score"] = np.nan
df.loc[(df["Sleep_Hours"] < 0) | (df["Sleep_Hours"] > 24), "Sleep_Hours"] = np.nan

df["Attendance"] = df["Attendance"].fillna(df["Attendance"].median())
df["Exam_Score"] = df["Exam_Score"].fillna(df["Exam_Score"].median())
df["Sleep_Hours"] = df["Sleep_Hours"].fillna(df["Sleep_Hours"].median())

print("----- Out-of-Range Values Fixed -----")

remaining_outliers = df[
    (df["Attendance"] < 0) | (df["Attendance"] > 100) |
    (df["Exam_Score"] < 0) | (df["Exam_Score"] > 100) |
    (df["Sleep_Hours"] < 0) | (df["Sleep_Hours"] > 24)
]

print("Remaining out-of-range rows:", remaining_outliers.shape[0])

display(df.loc[[5, 12, 20], ["Attendance", "Exam_Score", "Sleep_Hours"]])
```

```python
# 3. Fix unrecognized categories
df["Internet_Access"] = df["Internet_Access"].replace({
    "yup": "Yes",
    "maybe": np.nan
})
df["Gender"] = df["Gender"].replace({
    "M/F": np.nan
})
df["School_Type"] = df["School_Type"].replace({
    "unknown type": np.nan
})
df["Internet_Access"] = df["Internet_Access"].fillna(df["Internet_Access"].mode()[0])
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
df["School_Type"] = df["School_Type"].fillna(df["School_Type"].mode()[0])
print("----- Unrecognized Categories Fixed -----")
print("Internet_Access values:")
print(df["Internet_Access"].value_counts())
print("\nGender values:")
print(df["Gender"].value_counts())
print("\nSchool_Type values:")
print(df["School_Type"].value_counts())
display(df.loc[[3, 18, 25, 30], ["Internet_Access", "Gender", "School_Type"]])
```

```python
# 4. Fix duplicate rows

df = df.drop_duplicates()

print("----- Duplicate Rows Fixed -----")
print("Remaining duplicate rows:", df.duplicated().sum())
print("Dataset shape after removing duplicates:", df.shape)
```

```python
# Final after-cleaning evidence

print("----- After Cleaning: Missing Values -----")
print(df.isnull().sum())

print("\n----- After Cleaning: Duplicate Rows -----")
print(df.duplicated().sum())

print("\n----- Final Dataset Shape -----")
print(df.shape)
```

```python
# Scaling numeric columns using Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler
numeric_columns = ["Hours_Studied", "Attendance", "Sleep_Hours", "Previous_Scores", "Tutoring_Sessions", "Physical_Activity", "Exam_Score"]
df[numeric_columns] = MinMaxScaler().fit_transform(df[numeric_columns])
print("----- Min-Max Scaling Applied -----")
display(df[numeric_columns].head(10))
print("----- Minimum Values After Scaling -----")
print(df[numeric_columns].min())
print("----- Maximum Values After Scaling -----")
print(df[numeric_columns].max())
```
